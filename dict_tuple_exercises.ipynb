## ข้อ 6: Dictionary ที่มีค่าเป็น Tuple

สร้าง dictionary ชื่อ `students_scores` ที่เก็บข้อมูลนักเรียน:
```python
students_scores = {
    "นักเรียน1": (85, 90, 88),
    "นักเรียน2": (75, 80, 78),
    "นักเรียน3": (95, 92, 94)
}
```

ให้คำนวณค่าเฉลี่ยคะแนนของนักเรียนแต่ละคน และพิมพ์ผลลัพธ์

````python
# เขียนคำตอบของคุณที่นี่
students_scores = {
    "นักเรียน1": (85, 90, 88),
    "นักเรียน2": (75, 80, 78),
    "นักเรียน3": (95, 92, 94)
}

for name, scores in students_scores.items():
    average = sum(scores) / len(scores)
    print(f"{name} มีคะแนนเฉลี่ย = {average:.2f}")

````

## ข้อ 7: ค้นหาข้อมูลใน Dictionary

phone_book = {

## ข้อ 9: การรวม Dictionary

สร้าง dictionary 2 อัน:
- `person1 = {"name": "สมชาย", "age": 25}`
- `person2 = {"name": "สมหญิง", "age": 24}`

ให้รวม dictionary ทั้งสองเป็น dictionary เดียวชื่อ `all_people` และพิมพ์ผลลัพธ์

หมายเหตุ: ลองใช้วิธีต่างๆ เช่น update(), {**dict1, **dict2} เป็นต้น

````python
# เขียนคำตอบของคุณที่นี่
person1 = {"name": "สมชาย", "age": 25}
person2 = {"name": "สมหญิง", "age": 24}

all_people = person1.copy()
all_people.update(person2)

print(all_people)
````

## ข้อ 10: โปรแกรมจัดการสินค้า (Challenge)

สร้างโปรแกรมจัดการสินค้าที่มีความสามารถดังนี้:

1. มี dictionary ชื่อ `inventory` ที่เก็บข้อมูลสินค้า: ชื่อสินค้า, จำนวน, และราคา (เก็บในรูป tuple)
2. สร้างฟังก์ชน `add_product()` เพื่อเพิ่มสินค้าใหม่
3. สร้างฟังก์ชั่น `update_quantity()` เพื่ออัปเดตจำนวนสินค้า
4. สร้างฟังก์ชั่น `get_total_value()` เพื่อคำนวณมูลค่ารวมของสินค้า
5. สร้างฟังก์ชั่น `show_inventory()` เพื่อแสดงรายการสินค้าทั้งหมด

ตัวอย่างการใช้งาน:
```
Inventory: {'น้ำตาล': (10, 50), 'แป้งมัน': (5, 100)}
Total value: 1000 บาท
```

````python
# เขียนคำตอบของคุณที่นี่
inventory = {}


def add_product(name, quantity, price):
    """เพิ่มสินค้าใหม่ หรือแจ้งเตือนหากสินค้ามีอยู่แล้ว"""
    if name in inventory:
        print(f"สินค้า '{name}' มีอยู่แล้วในระบบ!")
    else:
        inventory[name] = (quantity, price)
        print(f"เพิ่มสินค้า '{name}' เรียบร้อย!")


def update_quantity(name, quantity):
    """อัปเดตจำนวนสินค้า"""
    if name in inventory:
        _, price = inventory[name]
        inventory[name] = (quantity, price)
        print(f"อัปเดตจำนวน '{name}' เป็น {quantity}")
    else:
        print(f"ไม่พบสินค้า '{name}' ในระบบ")


def get_total_value():
    """คำนวณมูลค่ารวมของสินค้าในระบบ"""
    total = 0
    for name, (qty, price) in inventory.items():
        total += qty * price
    return total


def show_inventory():
    """แสดงรายการสินค้า"""
    print("\n=== รายการสินค้าในคลัง ===")
    for name, (qty, price) in inventory.items():
        print(f"- {name}: จำนวน {qty}, ราคา {price} บาท")
    print("==========================\n")

add_product("น้ำตาล", 10, 50)
add_product("แป้งมัน", 5, 100)

# แสดงรายการ
show_inventory()

# คำนวณมูลค่ารวม
total = get_total_value()
print(f"Total value: {total} บาท")
````
