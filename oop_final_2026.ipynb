## ข้อที่ 1: การสร้างคลาส การเข้าถึง Attribute และ Method (20 คะแนน)

ให้สร้างคลาส `Student` ที่มีรายละเอียดดังนี้:

1. มี method `__init__` ที่รับพารามิเตอร์ `name` (ชื่อ), `student_id` (รหัสนักศึกษา) และ `gpa` (เกรดเฉลี่ย) แล้วเก็บเป็น attribute ของ object
2. มี method ชื่อ `show_info()` ที่แสดงข้อมูลในรูปแบบ:
   ```
   ชื่อ: <name> รหัส: <student_id> GPA: <gpa>
   ```
3. มี method ชื่อ `is_honor()` ที่คืนค่า `True` ถ้า GPA >= 3.25 และคืนค่า `False` ถ้า GPA < 3.25

จากนั้น:
- สร้าง object ชื่อ `s1` จากคลาส `Student` โดยกำหนด name="สมชาย", student_id="6501001", gpa=3.50
- สร้าง object ชื่อ `s2` จากคลาส `Student` โดยกำหนด name="สมหญิง", student_id="6501002", gpa=2.80
- เรียก method `show_info()` ของทั้ง `s1` และ `s2`
- แสดงผลว่านักศึกษาแต่ละคนได้เกียรตินิยมหรือไม่ โดยใช้ method `is_honor()`

````python
# เขียนโค้ดข้อที่ 1 ที่นี่
class Student:
    def  __init__(self, name, student_id, gpa):
        self.name = name
        self.student_id = student_id
        self.gpa = float(gpa)

    def show_info(self):
        print("{}: {} {}".format(self.name, self.student_id, self.gpa))
        
        
    def is_honor(self):
        return self.gpa >= 3.25
        
````

````python
s1 = Student("สมชาย", "6501001", "3.50")
s1.show_info()
````

````python
s2 = Student("สมหญิง", "6501002", "2.80")
s2.show_info()
````

````python
# === Test Cell ข้อที่ 1 (ห้ามแก้ไข) ===
print("--- ทดสอบข้อที่ 1 ---")
try:
    s1.show_info()
    s2.show_info()
    print(f"s1 ได้เกียรตินิยม: {s1.is_honor()}")
    print(f"s2 ได้เกียรตินิยม: {s2.is_honor()}")
    assert s1.name == "สมชาย"
    assert s1.is_honor() == True
    assert s2.is_honor() == False
    print("✅ ผ่านการทดสอบ")
except Exception as e:
    print(f"❌ ไม่ผ่าน: {e}")
````

## ข้อที่ 2: การ Encapsulation และการกำหนดค่า Attribute ผ่าน Method (20 คะแนน)

ให้สร้างคลาส `BankAccount` ที่มีรายละเอียดดังนี้:

1. มี method `__init__` ที่รับพารามิเตอร์ `owner` (ชื่อเจ้าของบัญชี) และ `balance` (ยอดเงินเริ่มต้น มีค่า default = 0)
2. เก็บ `owner` เป็น attribute แบบ public และเก็บ `balance` เป็น attribute แบบ **private** (ใช้ `__balance`)
3. มี method `deposit(amount)` สำหรับฝากเงิน โดย amount ต้องมากกว่า 0 จึงจะฝากได้ ถ้า amount <= 0 ให้ print "จำนวนเงินไม่ถูกต้อง"
4. มี method `withdraw(amount)` สำหรับถอนเงิน โดยถอนได้เมื่อ amount > 0 และ amount ไม่เกินยอดเงินคงเหลือ
   - ถ้า amount <= 0 ให้ print "จำนวนเงินไม่ถูกต้อง"
   - ถ้า amount > ยอดเงินคงเหลือ ให้ print "ยอดเงินไม่เพียงพอ"
5. มี method `get_balance()` ที่คืนค่ายอดเงินคงเหลือ

จากนั้น:
- สร้าง object ชื่อ `acc1` โดยกำหนด owner="สมศรี" และ balance=1000
- ฝากเงิน 500 บาท
- ถอนเงิน 200 บาท
- แสดงยอดเงินคงเหลือ

````python
# เขียนโค้ดข้อที่ 2 ที่นี่
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        else:
            print("จำนวนเงินไม่ถูกต้อง")

    def withdraw(self, amount):
        if amount <= 0:
            print("จำนวนเงินไม่ถูกต้อง")
        elif amount > self.__balance:
            print("ยอดเงินไม่เพียงพอ")
        else:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance

acc1 = BankAccount("สมชาย", 1000)
acc1.deposit(500)
acc1.withdraw(200)
print("ยอดเงินคงเหลือ:", acc1.get_balance())
````

````python
# === Test Cell ข้อที่ 2 (ห้ามแก้ไข) ===
print("--- ทดสอบข้อที่ 2 ---")
try:
    test_acc = BankAccount("ทดสอบ", 1000)
    test_acc.deposit(500)
    assert test_acc.get_balance() == 1500, f"ฝาก 500 แล้วต้องได้ 1500 แต่ได้ {test_acc.get_balance()}"
    test_acc.withdraw(200)
    assert test_acc.get_balance() == 1300, f"ถอน 200 แล้วต้องได้ 1300 แต่ได้ {test_acc.get_balance()}"
    test_acc.withdraw(5000)  # ควร print ยอดเงินไม่เพียงพอ
    assert test_acc.get_balance() == 1300, "ถอนเกินไม่ควรเปลี่ยนยอด"
    test_acc.deposit(-100)   # ควร print จำนวนเงินไม่ถูกต้อง
    # ทดสอบ private attribute
    try:
        _ = test_acc.__balance
        print("❌ __balance ไม่ควรเข้าถึงได้โดยตรง")
    except AttributeError:
        print("✅ __balance เป็น private attribute ถูกต้อง")
    print("✅ ผ่านการทดสอบ")
except Exception as e:
    print(f"❌ ไม่ผ่าน: {e}")
````

## ข้อที่ 3: การสืบทอดคลาส (Inheritance) อย่างง่าย (20 คะแนน)

ให้สร้างคลาสตามโครงสร้างดังนี้:

### คลาสแม่: `Animal`
1. มี method `__init__` ที่รับพารามิเตอร์ `name` (ชื่อสัตว์) และ `sound` (เสียงร้อง) แล้วเก็บเป็น attribute
2. มี method `speak()` ที่คืนค่า string ในรูปแบบ: `"<name> ร้อง <sound>"`
3. มี method `get_type()` ที่คืนค่า string `"สัตว์ทั่วไป"`

### คลาสลูก: `Dog` (สืบทอดจาก `Animal`)
1. มี method `__init__` ที่รับพารามิเตอร์ `name` และ `breed` (สายพันธุ์)
   - เรียกใช้ `__init__` ของคลาสแม่ โดยกำหนด sound เป็น `"โฮ่ง โฮ่ง"`
   - เก็บ `breed` เป็น attribute
2. มี method `fetch()` ที่คืนค่า string `"<name> กำลังวิ่งเก็บของ!"`

### คลาสลูก: `Cat` (สืบทอดจาก `Animal`)
1. มี method `__init__` ที่รับพารามิเตอร์ `name` และ `color` (สี)
   - เรียกใช้ `__init__` ของคลาสแม่ โดยกำหนด sound เป็น `"เหมียว"`
   - เก็บ `color` เป็น attribute
2. มี method `scratch()` ที่คืนค่า string `"<name> กำลังลับเล็บ!"`

จากนั้น:
- สร้าง object `d1` จากคลาส `Dog` โดย name="บัดดี้" และ breed="โกลเด้นรีทรีฟเวอร์"
- สร้าง object `c1` จากคลาส `Cat` โดย name="มิ้นท์" และ color="ขาว"
- เรียก method `speak()` ของทั้ง `d1` และ `c1`
- เรียก method `fetch()` ของ `d1` และ method `scratch()` ของ `c1`

````python
# เขียนโค้ดข้อที่ 3 ที่นี่
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound

    def speak(self):
        return f"{self.name} ร้อง {self.sound}"

    def get_type(self):
        return "สัตว์ทั่วไป"

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, "โฮ่ง โฮ่ง")
        self.breed = breed

    def fetch(self):
        return f"{self.name} กำลังวิ่งเก็บของ!"

class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name, "เหมียว")
        self.color = color

    def scratch(self):
        return f"{self.name} กำลังลับเล็บ!"

        
````

````python
d1 = Dog("บัดดี้", "โกลเด้นรีทรีฟเวอร์")
c1 = Cat("มิ้นท์", "ขาว")

print(d1.speak())
print(c1.speak())

print(d1.fetch())
print(c1.scratch())
````

````python
# === Test Cell ข้อที่ 3 (ห้ามแก้ไข) ===
print("--- ทดสอบข้อที่ 3 ---")
try:
    assert issubclass(Dog, Animal), "Dog ต้องสืบทอดจาก Animal"
    assert issubclass(Cat, Animal), "Cat ต้องสืบทอดจาก Animal"
    
    assert d1.speak() == "บัดดี้ ร้อง โฮ่ง โฮ่ง", f"speak() ไม่ถูกต้อง: {d1.speak()}"
    assert c1.speak() == "มิ้นท์ ร้อง เหมียว", f"speak() ไม่ถูกต้อง: {c1.speak()}"
    assert d1.fetch() == "บัดดี้ กำลังวิ่งเก็บของ!", f"fetch() ไม่ถูกต้อง: {d1.fetch()}"
    assert c1.scratch() == "มิ้นท์ กำลังลับเล็บ!", f"scratch() ไม่ถูกต้อง: {c1.scratch()}"
    assert d1.breed == "โกลเด้นรีทรีฟเวอร์"
    assert c1.color == "ขาว"
    assert d1.get_type() == "สัตว์ทั่วไป", "Dog ต้องสืบทอด get_type() จาก Animal"
    
    print(f"d1.speak() = {d1.speak()}")
    print(f"c1.speak() = {c1.speak()}")
    print(f"d1.fetch() = {d1.fetch()}")
    print(f"c1.scratch() = {c1.scratch()}")
    print("✅ ผ่านการทดสอบ")
except Exception as e:
    print(f"❌ ไม่ผ่าน: {e}")
````

## ข้อที่ 4: การทำ Method Overriding (20 คะแนน)

ให้สร้างคลาสตามโครงสร้างดังนี้:

### คลาสแม่: `Shape`
1. มี method `__init__` ที่รับพารามิเตอร์ `name` (ชื่อรูปร่าง) แล้วเก็บเป็น attribute
2. มี method `area()` ที่คืนค่า `0`
3. มี method `describe()` ที่คืนค่า string `"<name> มีพื้นที่ <area()> ตร.หน่วย"`

### คลาสลูก: `Rectangle` (สืบทอดจาก `Shape`)
1. มี method `__init__` ที่รับพารามิเตอร์ `width` (กว้าง) และ `height` (สูง)
   - เรียกใช้ `__init__` ของคลาสแม่ โดยกำหนด name เป็น `"สี่เหลี่ยม"`
   - เก็บ `width` และ `height` เป็น attribute
2. **Override** method `area()` ให้คืนค่า `width * height`

### คลาสลูก: `Circle` (สืบทอดจาก `Shape`)
1. มี method `__init__` ที่รับพารามิเตอร์ `radius` (รัศมี)
   - เรียกใช้ `__init__` ของคลาสแม่ โดยกำหนด name เป็น `"วงกลม"`
   - เก็บ `radius` เป็น attribute
2. **Override** method `area()` ให้คืนค่า `3.14159 * radius * radius` (ใช้ค่า π = 3.14159)

จากนั้น:
- สร้าง object `r1` จากคลาส `Rectangle` โดย width=5, height=10
- สร้าง object `c1` จากคลาส `Circle` โดย radius=7
- แสดงผล `describe()` ของทั้ง `r1` และ `c1`
- สร้าง list ชื่อ `shapes` ที่เก็บ `r1` และ `c1` แล้ววนลูป แสดงผล `describe()` ของแต่ละตัว

````python
# เขียนโค้ดข้อที่ 4 ที่นี่
class Shape:
    def __init__(self, name):
        self.name = name

    def area(self):
        return 0
        
    def describe(self):
        return f"{self.name} มีพื้นที่ {self.area()} ตร.หน่วย"

class Rectangle(Shape):
    def __init__(self, width, height):
        super().__init__("สี่เหลี่ยม")
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        super().__init__("วงกลม")
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius * self.radius
````

````python
r1 = Rectangle(5, 10)
c1 = Circle(7)

print("r1.describe() =", r1.describe())
print("c1.describe() =", c1.describe())

shapes = [r1, c1]

for shape in shapes:
    print(shape.describe())
````

````python
# === Test Cell ข้อที่ 4 (ห้ามแก้ไข) ===
print("--- ทดสอบข้อที่ 4 ---")
try:
    assert issubclass(Rectangle, Shape), "Rectangle ต้องสืบทอดจาก Shape"
    assert issubclass(Circle, Shape), "Circle ต้องสืบทอดจาก Shape"
    
    assert r1.area() == 50, f"Rectangle area ต้องเป็น 50 แต่ได้ {r1.area()}"
    assert abs(c1.area() - 153.93791) < 0.01, f"Circle area ไม่ถูกต้อง: {c1.area()}"
    
    print(f"r1.describe() = {r1.describe()}")
    print(f"c1.describe() = {c1.describe()}")
    
    # ทดสอบ Polymorphism ผ่าน list
    assert len(shapes) == 2, "shapes ต้องมี 2 ตัว"
    for s in shapes:
        assert isinstance(s, Shape), "ทุกตัวใน shapes ต้องเป็น Shape"
        print(f"  > {s.describe()}")
    
    # ทดสอบว่า method area() ถูก override จริง
    base = Shape("ทดสอบ")
    assert base.area() == 0, "Shape.area() ต้องคืน 0"
    
    print("✅ ผ่านการทดสอบ")
except Exception as e:
    print(f"❌ ไม่ผ่าน: {e}")
````

## ข้อที่ 5: การสืบทอดคลาสและ Method Overriding — ระบบสินค้า (20 คะแนน)

ให้สร้างคลาสตามโครงสร้างดังนี้:

### คลาสแม่: `Product`
1. มี method `__init__` ที่รับพารามิเตอร์ `name` (ชื่อสินค้า) และ `price` (ราคา) แล้วเก็บเป็น attribute
2. มี method `get_discount()` ที่คืนค่า `0` (ไม่มีส่วนลด)
3. มี method `final_price()` ที่คืนค่า `price - get_discount()`

### คลาสลูก: `Food` (สืบทอดจาก `Product`)
1. มี method `__init__` ที่รับพารามิเตอร์ `name` และ `price`
   - เรียกใช้ `__init__` ของคลาสแม่
2. **Override** method `get_discount()` ให้คืนค่า `price * 0.05` (ลด 5%)

### คลาสลูก: `Electronic` (สืบทอดจาก `Product`)
1. มี method `__init__` ที่รับพารามิเตอร์ `name` และ `price`
   - เรียกใช้ `__init__` ของคลาสแม่
2. **Override** method `get_discount()` ให้คืนค่า `price * 0.15` (ลด 15%)

จากนั้น:
- สร้าง object `p1` จากคลาส `Food` โดย name="ข้าวกล่อง" และ price=50
- สร้าง object `p2` จากคลาส `Electronic` โดย name="หูฟัง" และ price=1000
- แสดงผล `final_price()` ของทั้ง `p1` และ `p2`

````python
# เขียนโค้ดข้อที่ 5 ที่นี่
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
        
    def get_discount(self):
        return 0 

    def final_price(self):
        return self.price - self.get_discount()

class Food(Product):
    def __init__(self, name, price):
        super().__init__(name, price)

    def get_discount(self):
        return self.price * 0.05

class Electronic(Product):
    def __init__(self, name, price):
        super().__init__(name, price)

    def get_discount(self):
        return self.price * 0.15
        
````

````python
p1 = Food("ข้าวกล่อง", 50)
p2 = Electronic("หูฟัง", 1000)

print("ราคาสุทธิ", p1.name, "=", p1.final_price())
print("ราคาสุทธิ", p2.name, "=", p2.final_price())
````

````python
# === Test Cell ข้อที่ 5 (ห้ามแก้ไข) ===
print("--- ทดสอบข้อที่ 5 ---")
try:
    assert issubclass(Food, Product), "Food ต้องสืบทอดจาก Product"
    assert issubclass(Electronic, Product), "Electronic ต้องสืบทอดจาก Product"
    
    # ทดสอบ Food
    assert p1.name == "ข้าวกล่อง", f"ชื่อสินค้าไม่ถูกต้อง: {p1.name}"
    assert p1.get_discount() == 2.5, f"Food discount ต้องเป็น 2.5 แต่ได้ {p1.get_discount()}"
    assert p1.final_price() == 47.5, f"Food final_price ต้องเป็น 47.5 แต่ได้ {p1.final_price()}"
    
    # ทดสอบ Electronic
    assert p2.name == "หูฟัง", f"ชื่อสินค้าไม่ถูกต้อง: {p2.name}"
    assert p2.get_discount() == 150.0, f"Electronic discount ต้องเป็น 150.0 แต่ได้ {p2.get_discount()}"
    assert p2.final_price() == 850.0, f"Electronic final_price ต้องเป็น 850.0 แต่ได้ {p2.final_price()}"
    
    # ทดสอบว่า Product base ไม่มีส่วนลด
    base = Product("ทดสอบ", 100)
    assert base.get_discount() == 0, "Product.get_discount() ต้องคืน 0"
    assert base.final_price() == 100, "Product.final_price() ต้องคืน 100"
    
    print(f"p1: {p1.name} ราคา {p1.price} ส่วนลด {p1.get_discount()} ราคาสุทธิ {p1.final_price()}")
    print(f"p2: {p2.name} ราคา {p2.price} ส่วนลด {p2.get_discount()} ราคาสุทธิ {p2.final_price()}")
    print("✅ ผ่านการทดสอบ")
except Exception as e:
    print(f"❌ ไม่ผ่าน: {e}")
````
